Загрузите модуль pyplot библиотеки Matplotlib с псевдонимом plt, а также библиотеку Numpy с псевдонимом np.
Примените магическую функцию %matplotlib inline для отображения графиков в Jupyter Notebook и настройки конфигурации ноутбука со значением 'svg' для более четкого отображения графиков.

Постройте линейный график с помощью функции plot, отображающий зависимость квадрата числа от самого числа на интервале от -10 до 10.

```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

x = np.linspace(-10, 10, 100)
y = x ** 2

plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('x^2')
plt.title('График зависимости x^2 от x')

plt.grid()
plt.show()
```

Постройте на одном графике синусоиду и косинусоиду от 0 до

. Используйте разные цвета и стили линий для их отображения. С помощью функции linspace из библиотеки Numpy создайте массив t из 51 числа от 0 до 10 включительно.
Создайте массив Numpy под названием f, содержащий косинусы и синусы элементов массива t.
Постройте линейную диаграмму, используя массив t для координат по горизонтали, а массив f - для координат по вертикали. Определите для линии данного графика зеленый цвет.
Выведите название диаграммы - 'График f(t)'.
Также добавьте названия для горизонтальной оси - 'Значения t' и для вертикальной - 'Значения f'.
Ограничьте график по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.

```python
t = np.linspace(0, 10, 51)

f = np.sin(t) + np.cos(t)

plt.plot(t, np.sin(t), linestyle='--', label='sin(t)')
plt.plot(t, np.cos(t), linestyle=':', label='cos(t)')

plt.plot(t, f, color='green', label='f(t) = sin(t) + cos(t)')

plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')

plt.xlim(0.5, 9.5)
plt.ylim(-2.5, 2.5)

plt.grid()
plt.legend()
plt.show()
```

Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig и массив объектов Axes под названием ax, причем так, чтобы было 2 отдельных графика в сетке, состоящей из одной строки и двух столбцов. В каждом графике массив x используется для координат по горизонтали.

```python
x = np.linspace(-3, 3, 51)

fig, ax = plt.subplots(1, 2)

ax[0].plot(x, np.exp(x))
ax[0].set_title('Экспонента e^x')
ax[0].set_xlabel('x')
ax[0].set_ylabel('e^x')
ax[0].grid()

x_log = x[x > 0]

ax[1].plot(x_log, np.log(x_log))
ax[1].set_title('Натуральный логарифм ln(x)')
ax[1].set_xlabel('x')
ax[1].set_ylabel('ln(x)')
ax[1].grid()

plt.show()
```

На следующем графике постройте две гистограммы по значениям признака V1 - одну для мошеннических транзакций (Class равен 1) и другую - для обычных (Class равен 0). Подберите значение аргумента density так, чтобы по вертикали графика было расположено не число наблюдений, а плотность распределения. Число бинов должно равняться 20 для обеих гистограмм, а коэффициент alpha примите равным 0.5, чтобы гистограммы были полупрозрачными и не загораживали друг друга. Создайте легенду с двумя значениями: “Class 0” и “Class 1”. Гистограмма обычных транзакций должна быть серого цвета, а мошеннических - красного. Горизонтальной оси дайте название “V1”.

Постройте круговую диаграмму, отображающую доли мошеннических  (Class равен 1) и обычных транзакций (Class равен 0).

```python
plt.style.use('fivethirtyeight')

df = pd.read_csv('creditcard.csv')

plt.figure(figsize=(10, 6))
plt.hist(df['Time'], bins=50)
plt.title('Распределение транзакций по времени')
plt.xlabel('Time')
plt.ylabel('Количество')
plt.grid()
plt.show()


class_counts = df['Class'].value_counts()
plt.figure(figsize=(10, 6))
class_counts.plot(kind='bar')
plt.title('Количество транзакций по классам')
plt.xlabel('Class')
plt.ylabel('Количество')
plt.grid(axis='y')
plt.show()

plt.figure(figsize=(10, 6))
class_counts.plot(kind='bar', logy=True)
plt.title('Количество транзакций по классам (лог. масштаб)')
plt.xlabel('Class')
plt.ylabel('Количество (log)')
plt.grid(axis='y')
plt.show()


plt.figure(figsize=(10, 6))

plt.hist(
    df[df['Class'] == 0]['V1'],
    bins=20,
    density=True,
    alpha=0.5,
    color='gray',
    label='Class 0'
)

plt.hist(
    df[df['Class'] == 1]['V1'],
    bins=20,
    density=True,
    alpha=0.5,
    color='red',
    label='Class 1'
)

plt.title('Распределение V1 по классам')
plt.xlabel('V1')
plt.ylabel('Плотность')
plt.legend()
plt.grid()
plt.show()


plt.figure(figsize=(10, 6))

plt.pie(
    class_counts,
    labels=['Class 0', 'Class 1'],
    autopct='%1.2f%%',
    startangle=90
)

plt.title('Доли транзакций')
plt.axis('equal')

plt.show()
```

Используйте набор данных tips из библиотеки Seaborn. Постройте график распределения (distplot), который показывает распределение переменной total_bill. Добавьте кривую плотности распределения (KDE).
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
Постройте диаграмму рассеяния (scatterplot), показывающую зависимость между переменными total_bill и tip. Раскрасьте точки соответственно пола клиента (sex) и добавьте регрессионную линию.
Постройте диаграмму типа "ящик с усами" (boxplot), которая показывает распределение переменной total_bill в зависимости от дня недели (day). Используйте различную цветовую палитру для каждого дня.
Создайте сетку графиков с помощью FacetGrid, отображающую зависимость между переменными total_bill и tip в наборе данных tips с разбивкой по времени суток (time) и полу пассажира (sex). Внутри каждого графика постройте регрессионную линию.

```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

tips = sns.load_dataset('tips')

plt.figure(figsize=(10, 6))
sns.histplot(tips['total_bill'], kde=True)
plt.title('Распределение total_bill')
plt.xlabel('total_bill')
plt.ylabel('Количество')
plt.show()


plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
sns.regplot(data=tips, x='total_bill', y='tip', scatter=False, color='black')
plt.title('Зависимость чаевых от счета')
plt.show()


plt.figure(figsize=(10, 6))
sns.boxplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='day',
    palette='Set2',
    legend=False
)
plt.title('Распределение total_bill по дням')
plt.show()


g = sns.FacetGrid(tips, col='time', row='sex')

g.map_dataframe(sns.regplot, x='total_bill', y='tip', scatter_kws={'alpha': 0.5})

g.set_axis_labels('total_bill', 'tip')
g.fig.suptitle('FacetGrid: total_bill vs tip', y=1.02)
g.fig.set_size_inches(12, 8)
plt.show()
```

Задание 3

Используйте набор данных titanic из библиотеки Seaborn.
Используйте данные по корреляции признаков из набора данных titanic, чтобы построить тепловую карту (heatmap), которая показывает взаимосвязь между различными переменными. Добавьте аннотации (значения корреляций) на тепловую карту.
Постройте KDE-график, показывающий распределение переменной fare из набора данных titanic, и отобразите это распределение отдельно для каждого класса пассажиров (class).
Используйте метод lmplot, чтобы построить диаграмму, показывающую зависимость между переменными age и fare для пассажиров Титаника, с разбивкой по полу (sex) и классу обслуживания (class). Добавьте регрессионную линию.

```python
titanic = sns.load_dataset('titanic')


plt.figure(figsize=(10, 6))

corr = titanic.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Тепловая карта корреляций Titanic')
plt.show()


plt.figure(figsize=(10, 6))

sns.kdeplot(data=titanic, x='fare', hue='class', fill=True, common_norm=False, alpha=0.4)

plt.title('Распределение fare по классам пассажиров')
plt.xlabel('fare')
plt.ylabel('Плотность')
plt.grid(True, alpha=0.3)
plt.show()


g = sns.lmplot(
    data=titanic,
    x='age',
    y='fare',
    hue='sex',
    col='class',
    height=5,
    aspect=1.2
)

g.fig.suptitle('Зависимость age и fare по полу и классу', y=1.03)
plt.show()
```
